```python
import pandas as pd
```

```python
coord_df = pd.read_csv("countries_codes_and_coordinates.csv")
```

```python
coord_df.head()
```

```python
coord_df = coord_df.drop(['Alpha-2 code','Numeric code'], axis=1)
```

```python
coord_df.head()
```

```python
coord_df = coord_df.rename(columns={"Latitude (average)": "Latitude", "Longitude (average)": "Longitude", "Alpha-3 code": "NOC"})
```

```python
coord_df.columns = coord_df.columns.str.replace('"', '')
```

```python
coord_df['Latitude'] = coord_df['Latitude'].str.replace('"', '')
```

```python
coord_df['Longitude'] = coord_df['Longitude'].str.replace('"', '')
coord_df['NOC'] = coord_df['NOC'].str.replace('"', '')
```

```python
coord_df.head()
```

```python
coord_df['Latitude'] = coord_df['Latitude'].astype(float)
```

```python
coord_df['Longitude'] = coord_df['Longitude'].astype(float)
coord_df.head()
```
